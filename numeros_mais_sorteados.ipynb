```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

# Analise feita com base nos dados de 11/03/1996 até o concurso 2612 do dia 19/07/2023 <h5> [Fonte](https://asloterias.com.br/download-todos-resultados-mega-sena "asloterias.com.br")</h5>

```python
numbers = pd.read_excel('mega_sena_asloterias_ate_concurso_2612_crescente.xlsx', index_col=0)
```

```python
numbers.isna().sum()
```

```python
numbers = numbers.dropna()
```

```python
numbers.isna().sum()
```

```python
numbers = numbers.rename(columns={ 'Unnamed: 1':'data','Unnamed: 2':'b1', 'Unnamed: 3':'b2', 'Unnamed: 4':'b3', 'Unnamed: 5':'b4', 'Unnamed: 6':'b5',
                                            'Unnamed: 7':'b6'})                                         
```

```python
numbers = numbers[1:]
```

```python
index_rename = 'concurso'
numbers = numbers.rename_axis(index_rename)
```

```python
numbers.head(10)
```

```python
numbers.tail(10)
```

```python
obj_to_int = {'b1':int, 'b2':int, 'b3':int, 'b4':int, 'b5':int, 'b6':int}
numbers = numbers.astype(obj_to_int)
numbers.info()
```

```python
all_numbers = pd.concat([numbers['b1'], numbers['b2'], numbers['b3'], numbers['b4'], numbers['b5'],
                                 numbers['b6']], axis=0).reset_index(drop=True)
```

```python
all_numbers
```

```python
chart_height = all_numbers.value_counts().count()
chart_height
```

```python
plt.figure(figsize=(16, 6))
all_numbers_plt = all_numbers.value_counts().plot.bar(width=0.9)
for p in all_numbers_plt.patches:
    all_numbers_plt.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom', rotation=45)
    plt.title('Números que saem com mais frequência em sorteios da Mega Sena')
plt.xlabel('Números sorteados')
plt.ylabel('Frequência')
plt.axis([-1, chart_height , 0, 350])
plt.xticks(rotation=0)
plt.show()
```

```python
percent = (all_numbers.value_counts() / all_numbers.count() * 100).round(2)
```

```python
plt.figure(figsize=(16, 2))
all_numbers_plt = percent.plot.bar(width=0.9)
for p in all_numbers_plt.patches:
    all_numbers_plt.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom', rotation=45, fontsize='small')
    plt.title('Números que saem com mais frequência em sorteios da Mega Sena')
plt.xlabel('Números sorteados')
plt.ylabel('Frequência %')
plt.axis([-1, chart_height , 0, 3])
plt.xticks(rotation=0)
plt.show()
```
