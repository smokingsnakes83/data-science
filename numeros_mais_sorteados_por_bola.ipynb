# Frequência de números em cada bola sorteada

```python
bola_1 = mega['bola: 1'].value_counts()
```

```python
plt.figure(figsize=(15, 5))
chart = bola_1.plot.bar(width=0.9)
for p in chart.patches:
    chart.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom', fontsize='small')
    plt.title('Frequencia de números sorteados na bola: 1')
plt.axis([-1, len(bola_1), 0, bola_1.max() + 15])
plt.ylabel('Frequência')
plt.xlabel('Números sorteados')
plt.xticks(rotation=0)
plt.show()
```
